### **Description**

The data used in the study was collected from a randomly selected sample of mobile phone users in the six Class 1 metropolises of India— Bengaluru, Chennai, Delhi, Kolkata, Mumbai and Hyderabad. These cities had seen a high number of active cases (22.30% of all cases), and deaths due to COVID-19 (37.99% of all cases), and the shortage of medical facilities. Respondents were residents of the city, aged between 21 and 60 years—randomly selected from a list of mobile telephone users. The survey was undertaken during the months of August and September 2020 (when the lockdown had ended, but COVID cases and deaths were rising rapidly) and then resurveyed again in July-August 2021 (after the second wave had ended). These periods are referred to as T1 and T2, respectively.

The survey was administered using the abridged shorter version of the DASS questionnaire, DASS-21 (Lovibond & Lovibond, 1995). The raw scores were used to create z-scores, which were then recoded into five categories (normal, mild, moderate, severe and extremely severe) following the methodology suggested by Lovibond and Lovibond (1995).

####**DASS-21 (Depression Anxiety Stress Scale – 21 items)** is a widely used psychological assessment tool developed by Peter Lovibond and Sue Lovibond. It is a shortened version of the original 42-item DASS and is designed to measure three related negative emotional states: Depression, Anxiety, and Stress.

The scale consists of 21 self-report items, with 7 items allocated to each subscale. Respondents rate how much each statement applied to them over the past week using a 4-point Likert scale ranging from 0 (“Did not apply to me at all”) to 3 (“Applied to me very much or most of the time”).

Scores for each subscale are calculated by summing the relevant items (and often multiplying by two to maintain comparability with the original DASS-42). Higher scores indicate greater levels of psychological distress. The DASS-21 is widely used in research and clinical settings due to its reliability, validity, and ability to distinguish between symptoms of depression, anxiety, and stress.

**Possible Solution:**

As economic status is directly related to education of child, so if we observe that a certain family's economic condition from phase T1 to T2,
* got better, we could assign `not affected` in T2
* if it got worse and the change is small and the family is in economically well enough, we could assign `not affected` in T2
* if it got worse and the change cannot be considered as small although not a drastic change either we could assign `somewhat affected` in T2
* if it got worse and the change is drastic, we could assign `severely affected` in T2

If economic status of a certain family in phase T1 is,
* better, we could assign `not affected`
* somewhat better, we could assign `somewhat affected`
* not well enogh, we could assign `severely affected`

###**Missing values in `monthly_expenditure T2`:**

```python
b = (mh['monthly_expenditure T1'] <= mh['monthly_expenditure T2']) & ((mh['education_of_children T1'].notna())) & (mh['education_of_children T2'].isna()) & (mh['mrd_and_spouse_living_together_or_not']!='Unmarried')
mh[b]
```

#####
Respondents whose `monthly_expenditure T2` was greater than or equal to T1 and who were married, but had missing values in `education_of_children T2` despite having valid responses in `education_of_children T1`, were assigned the category `somewhat affecte` for `education_of_children T2`.
* This imputation was performed using a mode-based approach, as`somewhat affecte` was the most frequently occurring category in the dataset. Given the relative stability or increase in household expenditure across phases, this category was considered the most statistically and contextually appropriate representation for the missing responses.

```python
mh.loc[b, 'education_of_children T2'] = mh.loc[b, 'education_of_children T2'].fillna('Somewhat affected')
mh[b]
```

###**Missing values in `monthly_expenditure T1`:**

```python
a = ((mh['education_of_children T1'].isna()) & (mh['education_of_children T2'].notna()) & (mh['mrd_and_spouse_living_together_or_not']!='Unmarried'))
a1 = (a & (abs(mh['monthly_expenditure T1'] - mh['monthly_expenditure T2'])<= 2000))
mh[a1]
```

####
For respondents (married) with missing values in `education_of_children T1` but available values in `education_of_children T2`, the T1 values were imputed using the corresponding T2 values when the difference in monthly_expenditure between T1 and T2 was less than ₹2000
* This approach is based on the assumption that households whose monthly expenditure did not change substantially across the two phases, were likely to  remain in a almost similar economic status over a span of one year, and therefore were expected to maintain similar educational expenditure patterns for children.


```python
mh.loc[a1, 'education_of_children T1'] = mh.loc[a1, 'education_of_children T2']
mh[a1]
```
